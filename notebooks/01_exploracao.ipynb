# Projeto Olist - Análise Exploratória

## Objetivo

Realizar uma análise exploratória dos dados do marketplace Olist para compreender a estrutura das tabelas, identificar métricas iniciais de negócio e preparar os dados para análises posteriores.

```python
import pandas as pd
import numpy as np
import os

pd.set_option('display.max_columns', None)
```

```python
customers = pd.read_csv('../data/olist_customers_dataset.csv')
orders = pd.read_csv('../data/olist_orders_dataset.csv')
products = pd.read_csv('../data/olist_products_dataset.csv')
items = pd.read_csv('../data/olist_order_items_dataset.csv')
payments = pd.read_csv('../data/olist_order_payments_dataset.csv')
reviews = pd.read_csv('../data/olist_order_reviews_dataset.csv')
geolocation = pd.read_csv('../data/olist_geolocation_dataset.csv')
sellers = pd.read_csv('../data/olist_sellers_dataset.csv')
category_translation = pd.read_csv('../data/product_category_name_translation.csv')
```

```python
datasets = {
    'customers': customers,
    'orders': orders,
    'products': products,
    'items': items,
    'payments': payments,
    'reviews': reviews,
    'geolocation': geolocation,
    'sellers': sellers,
    'category_translation': category_translation
}

for nome, df in datasets.items():
    print(nome)
    print(df.shape)
    print()
```

```python
orders.dtypes
orders['order_purchase_timestamp'].head()

lista_de_datas = [
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_purchase_timestamp',
    'order_approved_at'
]

# Converter as colunas listadas para datetime
for col in lista_de_datas:
    orders[col] = pd.to_datetime(orders[col], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# Verificar os tipos agora
orders[lista_de_datas].dtypes
datasets = {
    'customers': customers,
    'orders': orders,
    'products': products,
    'items': items,
    'payments': payments,
    'reviews': reviews,
    'geolocation': geolocation,
    'sellers': sellers,
    'category_translation': category_translation
}

for nome, df in datasets.items():
    print(nome)
    print(df.shape)
    print()
```
